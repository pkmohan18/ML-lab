```python
import csv
import math
def major_class(data, attributes, target):
    global MC
    MC=MC+1
    print("In Major_CLASS itereation:",MC)
    freq = {}
    index = attributes.index(target)
    #print("major_class index is {}".format(index))
    for t in data:
        if t[index] in freq:
            freq[t[index]] += 1
        else:
            freq[t[index]] = 1
    m = 0
    major = ""
    for key in freq.keys():
        if freq[key] > m:
            m = freq[key]
            major = key
    print("Returning major class: ",major)
    return major

def entropy(attributes, data, targetAttr):
    global EN
    EN=EN+1
    print("Entropy itereation:",EN)
    freq = {}
    data_entropy = 0.0
    i = 0
    for entry in attributes:
        if targetAttr == entry:
            break
        i += 1
    for entry in data:
        if entry[i] == 'PlayTennis':
            pass
        else:
            if entry[i] in freq:
                freq[entry[i]] += 1.0
            else:
                freq[entry[i]] = 1.0
    for f in freq.values():
        data_entropy += (-f/len(data)) * math.log(f/len(data), 2)
    print("Returning Entropy: ",data_entropy)
    return data_entropy

def info_gain(data, attributes, targetAttr, attr):
    global IG
    IG=IG+1
    print()
    print("INFO GAIN itereation:",IG)
    freq = {}
    subset_entropy = 0.0
    i = attributes.index(attr)
    for entry in data:
        if entry[i] == attr:
            pass
        else:
            if entry[i] in freq:
                freq[entry[i]] += 1.0
            else:
                freq[entry[i]] = 1
    for val in freq.keys():
        p = sum(freq.values())
        val_prob = freq[val] / (p)
        data_subset = [entry for entry in data if entry[i] == val]
        subset_entropy += val_prob * entropy(attributes, data_subset, targetAttr)
    data_subset = [entry for entry in data if entry[0] != 'Outlook']
    Info_gain=entropy(attributes, data_subset, targetAttr) - subset_entropy
    print("Returning Info Gain is: ",Info_gain)
    return Info_gain

def attr_choose(data, attributes, target):
    global AC
    AC=AC+1
    print()
    print("Attibute Choose itereation: ",AC)
    best = attributes[0]
    max_gain = 0
    for attr in attributes:
        if attr != target:
            new_gain = info_gain(data, attributes, target, attr)
            if new_gain > max_gain:
                max_gain = new_gain
                best = attr
    print("Best attribute Returned is: ",best)
    print()
    return best

def get_values(data, attributes, attr):
    print()
    print("I am in GET Values***")
    i = attributes.index(attr)
    values = []
    for entry in data:
        if entry[i] == attr:
            pass
        else:
            if entry[i] not in values:
                values.append(entry[i])
    print("Returning from GET_Values: ", values)
    print()
    return values

def get_data(data, attributes, best, val):
    print()
    print("I am in GET_DATA: ")
    print("Best attribute  and its value is: ", best ,val)
    print("Attributes has: ",attributes)
    new_data = []
    i = attributes.index(best)
    print("i value is: ",i)
    print()
    for entry in data:
        if entry[i] == val:
            new_entry = []
            for j in range(len(entry)):
                if j != i:
                    new_entry.append(entry[j])
                    print("Appending: ",entry[j],"to new entry: ",new_entry)
            new_data.append(new_entry)
    print("New_DATA Returning from GET_DATA: ",new_data)
    print()
    return new_data

def build_tree(data, attributes, target):
    global BT
    BT=BT+1
    print("BUILD TREE:",BT)
    print("Data contents:")
    print(data)    
    print("Attributes are:")
    print(attributes)
    print("Target Value is: ",target)
    print("Index Value is: ",attributes.index(target))
    print("Attribute at index is:",attributes[attributes.index(target)])
    vals = [record[attributes.index(target)] for record in data]
    print("New Dataset for this iteration is: ",data)
    print("New attributes for this iteration is: ",attributes)
    print("VALS ARE: ",vals)
    default = major_class(data, attributes, target)
    if not data or (len(attributes) - 1) <= 0:
        print("LENGHT OF THE ATTRIBUTE IS == 0")
        return default
    elif vals.count(vals[0]) == len(vals):
        print("Elseif: ",vals.count(vals[0]),"==", len(vals)," is true")
        print("Vals has",vals)
        print("returning: ",vals[0])
        return vals[0]
    else:
        best = attr_choose(data, attributes, target)
        tree = {best: {}}
        for val in get_values(data, attributes, best):
            new_data = get_data(data, attributes, best, val)
            new_attr = attributes[:]
            new_attr.remove(best)
            print("**************target is:", target)
            subtree = build_tree(new_data, new_attr, target)
            tree[best][val] = subtree
    return tree

def test(attributes, instance, tree):
    attribute = next(iter(tree))
    print("Attribute value is: ",attribute)
    i = attributes.index(attribute)
    print("Index value is: ",i)
    if instance[i] in tree[attribute].keys():
        result = tree[attribute][instance[i]]
        print("Result is: ",result)
        if isinstance(result, dict):
            print("Result ",result,", is of type dictionary.")
            print("Returning: test(attributes,instance,result) to test the subtree of dictionary.")
            return test(attributes, instance, result)
        else:
            print("Else Return Result: ",result)
            return result
    else:
        return 'NULL'
    
def execute_decision_tree():
    data = []
    with open('PlayTennis.csv') as tsv:
        for line in csv.reader(tsv):
            data.append(tuple(line))
        print('Number of records: ', len(data))
        print()
    
    attributes = list(data[0])
    target = attributes[-1]
    training_set = [x for i, x in enumerate(data)]    
    print("DATA SET IS:")
    for sample in training_set:
        print(sample)
    print("Training_set has:")
    print(training_set)
    print("Target has:")
    print(target)
    print()

    tree = build_tree(training_set, attributes, target)

    print()
    print("***************")
    print('Decision Tree is as below: \n')
    print(tree)
    instance = ['Rainy','Hot','Normal','False']

    print()
    print("***************")
    print('Testing instance is: ', instance)
    result = test(attributes, instance, tree)
    print('The Target value for the testing instance is: ')
    print(result)
#global BT
BT=AC=EN=IG=MC=0

execute_decision_tree()
```
